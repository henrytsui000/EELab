# EXP 4

```python
def moving_average(a, n=100) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
x_real = np.linspace(0, 2, 1001)
y_in = np.sin(10*x_real)+0.3*np.sin(200*x_real)*np.sin(10*x_real)
x_out, y_out = moving_average(x_real), -moving_average(y_in)
plt.plot(x_real, y_in)
plt.plot(x_out, y_out, linewidth=3)
plt.show()
```
